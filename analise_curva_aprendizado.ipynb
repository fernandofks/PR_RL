# Lunar Lander

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")

attr = ['time/total_timesteps','time/time_elapsed','rollout/ep_rew_mean']

lunar_lander_01_1 = pd.read_csv('results/lunar_lander_0.1_1/progress.csv')
lunar_lander_01_1 = lunar_lander_01_1[attr]

lunar_lander_01_10 = pd.read_csv('results/lunar_lander_0.1_10/progress.csv')
lunar_lander_01_10 = lunar_lander_01_10[attr]

lunar_lander_01_100 = pd.read_csv('results/lunar_lander_0.1_100/progress.csv')
lunar_lander_01_100 = lunar_lander_01_100[attr]

lunar_lander_05_1 = pd.read_csv('results/lunar_lander_0.5_1/progress.csv')
lunar_lander_05_1 = lunar_lander_05_1[attr]

lunar_lander_05_10 = pd.read_csv('results/lunar_lander_0.5_10/progress.csv')
lunar_lander_05_10 = lunar_lander_05_10[attr]

lunar_lander_05_100 = pd.read_csv('results/lunar_lander_0.5_100/progress.csv')
lunar_lander_05_100 = lunar_lander_05_100[attr]

lunar_lander_1_1 = pd.read_csv('results/lunar_lander_1.0_1/progress.csv')
lunar_lander_1_1 = lunar_lander_1_1[attr]

lunar_lander_1_10 = pd.read_csv('results/lunar_lander_1.0_10/progress.csv')
lunar_lander_1_10 = lunar_lander_1_10[attr]

lunar_lander_1_100 = pd.read_csv('results/lunar_lander_1.0_100/progress.csv')
lunar_lander_1_100 = lunar_lander_1_100[attr]
```

```python
lunar_lander_01_1['algol'] = '01_1'
lunar_lander_01_10['algol'] = '01_10'  
lunar_lander_01_100['algol'] = '01_100'
lunar_lander_05_1['algol'] = '05_1'
lunar_lander_05_10['algol'] = '05_10'
lunar_lander_05_100['algol'] = '05_100'  
lunar_lander_1_1['algol'] = '1_1'
lunar_lander_1_10['algol'] = '1_10'
lunar_lander_1_100['algol'] = '1_100'

df_1 = pd.concat([
    lunar_lander_01_1,
    lunar_lander_05_1,
    lunar_lander_1_1
])

df_10 = pd.concat([
    lunar_lander_01_10,
    lunar_lander_05_10,
    lunar_lander_1_10,
])

df_100 = pd.concat([
    lunar_lander_01_100,
    lunar_lander_05_100,
    lunar_lander_1_100,
])

df = pd.concat([
    lunar_lander_01_1,
    lunar_lander_01_10,
    lunar_lander_01_100,
    lunar_lander_05_1,
    lunar_lander_05_10,
    lunar_lander_05_100,
    lunar_lander_1_1,
    lunar_lander_1_10,
    lunar_lander_1_100,
])
```

```python
sns.lineplot(x="time/total_timesteps", y="rollout/ep_rew_mean", hue='algol', data=df)
plt.xlabel('Timesteps')
plt.ylabel('Recompensa')
plt.title('Comparando Lunar Lander')
plt.legend(title='Algoritmo')
plt.show()
```

```python
sns.lineplot(x="time/total_timesteps", y="rollout/ep_rew_mean", hue='algol', data=df_1)
plt.xlabel('Timesteps')
plt.ylabel('Recompensa')
plt.title('Comparando Lunar Lander')
plt.legend(title='Algoritmo')
plt.show()
```

```python
sns.lineplot(x="time/total_timesteps", y="rollout/ep_rew_mean", hue='algol', data=df_10)
plt.xlabel('Timesteps')
plt.ylabel('Recompensa')
plt.title('Comparando Lunar Lander')
plt.legend(title='Algoritmo')
plt.show()
```

```python
sns.lineplot(x="time/total_timesteps", y="rollout/ep_rew_mean", hue='algol', data=df_100)
plt.xlabel('Timesteps')
plt.ylabel('Recompensa')
plt.title('Comparando Lunar Lander')
plt.legend(title='Algoritmo')
plt.show()
```
